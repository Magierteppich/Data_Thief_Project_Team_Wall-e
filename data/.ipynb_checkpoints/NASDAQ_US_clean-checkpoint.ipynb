```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("NASDAQ_US.csv")
```

```python
df["date"]=pd.to_datetime(df["date"])
df
```

```python
columns = ["date","NASDAQ_US_value"]
df.columns = columns
df = (df
       .loc[(df["date"]>="2019-01-01")&(df["date"]<"2020-04-08"),:]
       .sort_values("date", ascending=False)
       .reset_index()
       .drop(columns="index")
        )
```

```python
df
```

```python
df.to_csv("NASDAQ_clean")
```
