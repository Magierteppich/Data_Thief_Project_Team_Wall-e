```python
#  Dual Plotting: Stock market index & new cases of Covid-19 over time (Germany)
x_values = df_stock_market_covid_19['date']
y1_values = df_stock_market_covid_19['DAX_30_GR_x']
y2_values = df_stock_market_covid_19['Germany']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('dates')
ax1.set_ylabel('DAX 30 normalized (DE)', color=color)
ax1.plot(x_values, y1_values, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('New COVID-19 cases daily, in Germany', color=color)  # we already handled the x-label with ax1
ax2.plot(x_values, y2_values, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
df = df_stock_market_covid_19.drop(columns=['SSE_CN', 'CAC_40_FR', 'DAX_30_GR', 'DOW_JONES_US', 'FTSE_IT', 'IBEX_SP','NIKKEI_JP', 'S&P_500_US', 'S&P_500_US_x'])
df = df.rename(columns={"CAC_40_FR_x":"CAC 40 normalized (FR)",
                        "DAX_30_GR_x":"DAX 30 normalized (DE)",
                        "DOW_JONES_US_x":"DOW JONES normalized (US)",
                        "FTSE_IT_x":"FTSE normalized (IT)",
                        "IBEX_SP_x":"IBEX 35 normalized (SP)",
                        "NIKKEI_JP_x":"NIKKEI normalized (JP)",
                        "SSE_CN_x": "SSE normalized (CN)",
                        "China":"Covid-19 cases (CN)",
                        "France":"Covid-19 cases (FR)",
                        "Germany":"Covid-19 cases (DE)",
                        "Italy":"Covid-19 cases (IT)",
                        "US":"Covid-19 cases (US)",
                        "Japan":"Covid-19 cases (JP)",
                        "Spain":"Covid-19 cases (SP)"})

# Splitting the main DataFrame into two DataFrames, i.e. for the stock market indexes and the number of cases of Covid-19
df_stockmarket = df[["date","CAC 40 normalized (FR)", "DAX 30 normalized (DE)", "DOW JONES normalized (US)", "FTSE normalized (IT)", "IBEX 35 normalized (SP)", "NIKKEI normalized (JP)", "SSE normalized (CN)"]]
df_covid19 = df[["date","Covid-19 cases (FR)", "Covid-19 cases (DE)", "Covid-19 cases (US)", "Covid-19 cases (IT)", "Covid-19 cases (SP)", "Covid-19 cases (JP)", "Covid-19 cases (CN)"]]
```

```python
#  Dual Plotting: Stock market index & new cases of Covid-19 over time (All)
fig, axes = plt.subplots(nrows=7,
                         ncols=1,
                         figsize=(7,30))

y1_stockmarket_list = ["CAC 40 normalized (FR)","DAX 30 normalized (DE)", "DOW JONES normalized (US)","FTSE normalized (IT)", "IBEX 35 normalized (SP)","NIKKEI normalized (JP)","SSE normalized (CN)"]
y2_covid19_list = ["Covid-19 cases (FR)", "Covid-19 cases (DE)", "Covid-19 cases (US)", "Covid-19 cases (IT)", "Covid-19 cases (SP)", "Covid-19 cases (JP)", "Covid-19 cases (CN)"]
color_stockmarket = ["blue", "yellow", "purple", "green", "orange", "brown", "red"]
color_covid19 = ["navy", "gold", "darkviolet", "darkgreen", "orangered", "salmon", "magenta"]

for i in range(7): 
    x_values = df['date']
    y1_values = df_stockmarket[y1_stockmarket_list[i]]
    y2_values = df_covid19[y2_covid19_list[i]]
    color = 'tab:red'
    axes[i].set_ylabel(y1_stockmarket_list[i], color=color, rotation=90)
    axes[i].plot(x_values, y1_values, color=color)
    axes[i].tick_params(axis='y', labelcolor=color)
    axes[i].grid(b=None)
    axes[i].set_ylim(bottom=0.6, top=1.1)
    axes[i] = axes[i].twinx()  
    color = 'tab:blue'
    axes[i].set_ylabel(y2_covid19_list[i], color=color, rotation=90)  # we already handled the x-label with ax1
    axes[i].plot(x_values, y2_values, color=color)
    axes[i].tick_params(axis='y', labelcolor=color)
    axes[i].grid(b=None)


plt.show()
```
